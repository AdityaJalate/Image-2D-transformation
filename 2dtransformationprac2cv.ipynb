```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
from google.colab.patches import cv2_imshow
# translation matrix
img=cv2.imread("car.jpeg")
tx,ty=50,10 #shift by 50 pixels to rigth abd 30 to left
translation_matrix=np.float32([[1,0,tx],[0,1,ty]])
img_translation=cv2.warpAffine(img,translation_matrix,(img.shape[1],img.shape[0]))
```

```python
cv2_imshow(img_translation)
```

```python
#rotation matrix
angle =40
center=(img.shape[0]/2,img.shape[1]/2)
rotation_matrix=cv2.getRotationMatrix2D(center,angle,1)
print(rotation_matrix)
rotation=cv2.warpAffine(img,rotation_matrix,(img.shape[1],img.shape[0]))
```

```python
cv2_imshow(rotation)
```

```python

from google.colab import files
#upload and load image
uploaded=files.upload()
image_path=list(uploaded.keys())[0]
img=cv2.cvtColor(cv2.imread(image_path),cv2.COLOR_BGR2RGB)

#show function
def show(img,title="image"):
  plt.imshow(img)
  plt.title(title)
  plt.show()

show(img,"original image")
```

```python
#transformation
rows,cols=img.shape[:2]

#translation
translation_matrix=np.float32([[1,0,tx],[0,1,ty]])
img_translation=cv2.warpAffine(img,translation_matrix,(cols,rows))
show(img_translation,"translation")
```

```python
#rotation
rotation_matrix=cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
img_rotation=cv2.warpAffine(img,rotation_matrix,(cols,rows))
show(img_rotation,"rotation")
```

```python
#scaling
scaled_up_iamge=cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)
show(scaled_up_iamge,"scaled up")
```

```python
# reflection
reflection_matrix=np.float32([[-1,0,cols],[0,1,0]])
img_reflection=cv2.warpAffine(img,reflection_matrix,(cols,rows))
show(img_reflection,"reflection")
```

```python
#shearing
shearing_matrix=np.float32([[1,0.5,0],[0.5,1,0]])
img_shearing=cv2.warpAffine(img,shearing_matrix,(cols,rows))
show(img_shearing,"shearing")
```
